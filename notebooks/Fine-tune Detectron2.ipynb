```python
from deepdoctection.datasets import get_dataset
from deepdoctection.eval import metric_registry
from deepdoctection.train.d2_frcnn_train import train_d2_faster_rcnn
```

```python
publaynet = get_dataset("publaynet")
path_config_yaml = "/home/janis/.cache/deepdoctection/configs/dd/d2/layout/CASCADE_RCNN_R_50_FPN_GN.yaml"
path_weights = "/home/janis/.cache/deepdoctection/weights/layout/d2_model-800000-layout.pkl"
build_train_config = ["max_datapoints=100000"]
dataset_train = publaynet
dataset_val = publaynet
build_val_config = ["max_datapoints=500"]
config_overwrite = ["SOLVER.IMS_PER_BATCH=2","SOLVER.BASE_LR=0.001","SOLVER.MAX_ITER=100000","SOLVER.STEPS=[]", "TEST.EVAL_PERIOD=10000"]
coco_metric = metric_registry.get("coco")

log_dir = "/home/janis/Documents/d2_training/layout"

train_d2_faster_rcnn(path_config_yaml,
                     dataset_train,
                     path_weights,
                     config_overwrite,
                     log_dir,
                     build_train_config,
                     dataset_val,
                     build_val_config,
                     None,
                     coco_metric,
                     "ImageLayoutService")
```

```python
pubtabnet = get_dataset("pubtabnet")
pubtabnet.dataflow.categories.filter_categories(categories="CELL")

path_config_yaml = "/home/janis/.cache/deepdoctection/configs/dd/d2/cell/CASCADE_RCNN_R_50_FPN_GN.yaml"
path_weights = "/home/janis/.cache/deepdoctection/weights/cell/d2_model-1800000-cell.pkl"

build_train_config = ["max_datapoints=100000"]
dataset_train = pubtabnet
dataset_val = pubtabnet

build_val_config = ["max_datapoints=500"]
config_overwrite = ["SOLVER.IMS_PER_BATCH=2","SOLVER.BASE_LR=0.001","SOLVER.MAX_ITER=100000","SOLVER.STEPS=[]", "TEST.EVAL_PERIOD=10000"]
coco_metric = metric_registry.get("coco")
coco_metric.set_params(max_detections=[50,200,600], area_range=[[0,1000000],[0,200],[200,800],[800,1000000]])

log_dir = "/home/janis/Documents/d2_training/cell"

train_d2_faster_rcnn(path_config_yaml,
                     dataset_train,
                     path_weights,
                     config_overwrite,
                     log_dir,
                     build_train_config,
                     dataset_val,
                     build_val_config,
                     None,
                     coco_metric,
                     "ImageLayoutService")
```
